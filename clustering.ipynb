```python
import pandas as pd
import numpy as np
import datetime
```

```python
df= pd.read_csv('^HSI.csv',index_col=0,parse_dates=True).dropna()['2001-07-09':]
```

```python

df=df.drop(columns=['Adj Close'])
```

```python
df
```

```python
df.shape
```

```python
max = 20
for i in range(0,max):
    for x in range(1,max-i):
        y=i+x
        # df[('N-'+str(i)+str(y)+' Open')]=df.shift(i)['Open']/df.shift(y)['Close']
        # df[('N-'+str(i)+str(y)+' High')]=df.shift(i)['High']/df.shift(y)['Close']
        # df[('N-'+str(i)+str(y)+' Low')]=df.shift(i)['Low']/df.shift(y)['Close']
        df[('N-'+str(i)+str(y)+' Close')]=df.shift(i)['Close']/df.shift(y)['Close']
        df[('N-'+str(i)+str(y)+' Volume')]=df.shift(i)['Volume']/df.shift(y)['Volume']
        df[('N-'+str(i)+str(y)+' Amplitude')]=(df.shift(i)['High']-df.shift(i)['Low'])/(df.shift(y)['High']-df.shift()['Low'])
```

```python
df['label']=(df.shift(-5)['Close']>df.shift(-1)['Open'])
```

```python
df['label'] = df['label'].astype(int)
```

```python
df.shape
```

```python
df=df.replace([np.inf, -np.inf], np.nan)
df=df.dropna().drop(columns=['Open','High','Low','Close','Volume'])
```

```python
df.shape
```

```python
df
```

```python
msk=np.random.rand(len(df))<0.9
train_data=df[msk]
test_data=df[~msk]
```

```python
from sklearn import preprocessing
```

```python
def process(data):
    ndarray=data.values
    label=ndarray[:,-1]
    features=ndarray[:,:-1]
    minmax_scale=preprocessing.MinMaxScaler(feature_range=(0,1))
    scaledFeatures=minmax_scale.fit_transform(features)
    return label, scaledFeatures
```

```python
train_label,train_features=process(train_data)
test_label,test_features=process(test_data)
```

```python
train_features
```

```python
train_label
```

```python
from keras.models import Sequential
from keras.layers import Dense,Dropout
```

```python
df.shape
```

```python
model=Sequential()
```

```python
model.add(Dense(units=5700,input_dim=570,kernel_initializer='uniform',activation='relu'))
```

```python
model.add(Dense(units=570,kernel_initializer='uniform',activation='relu'))
```

```python
model.add(Dense(units=57,kernel_initializer='uniform',activation='relu'))
```

```python
model.add(Dense(units=1,kernel_initializer='uniform',activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
train_history=model.fit(x=train_features,y=train_label,validation_split=0.1,epochs=1000,batch_size=4467,verbose=2)
```

```python
import matplotlib.pyplot as plt
def show_train_history(train_history,train,validation):
    plt.plot(train_history.history[train])
    plt.plot(train_history.history[validation])
    plt.title('Train History')
    plt.ylabel(train)
    plt.xlabel('Epoch')
    plt.legend(['train','validation'],loc='upper left')
    plt.show()
```

```python
show_train_history(train_history,'acc','val_acc')
```

```python
show_train_history(train_history,'loss','val_loss')
```

```python
scores=model.evaluate(x=test_features,y=test_label)
```
