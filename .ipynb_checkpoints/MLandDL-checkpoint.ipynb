```python
import pandas as pd
import datetime
```

```python
df= pd.read_csv('^HSI.csv',index_col=0,parse_dates=True).dropna()['2001-07-09':]
```

```python
df=df.drop(columns=['Adj Close'])
```

```python
for i in range(1,6):
    df[('N-'+str(i)+' Open')]=df.shift(i)['Open']/df['Open']
    df[('N-'+str(i)+' High')]=df.shift(i)['High']/df['Open']
    df[('N-'+str(i)+' Low')]=df.shift(i)['Low']/df['Open']
    df[('N-'+str(i)+' Close')]=df.shift(i)['Close']/df['Open']
    df[('N-'+str(i)+' Volume')]=df.shift(i)['Volume']/df['Volume']
```

```python
df=df.dropna()
```

```python
df['High']=df['High']/df['Open']
df['Low']=df['Low']/df['Open']
df['Close']=df['Close']/df['Open']
```
